```python
from skimage.transform import rotate
from skimage.feature import local_binary_pattern
from skimage import data
from skimage.color import label2rgb

import numpy as np
import matplotlib.pyplot as plt

METHOD = 'uniform'

# settings for LBP
radius = 1
n_points = 8 * radius


def overlay_labels(image, lbp, labels):
    mask = np.logical_or.reduce([lbp == each for each in labels])
    return label2rgb(mask, image=image, bg_label=0, alpha=0.5)


def highlight_bars(bars, indexes):
    for i in indexes:
        bars[i].set_facecolor('r')

import imageio.v3 as iio

def imread_grayscale(path):
    return np.asarray(Image.open(path).convert("L"), dtype=np.float32) / 255.0

image = imread_grayscale('img/lbp/img1.png')
lbp = local_binary_pattern(image, n_points, radius, METHOD)


def hist(ax, lbp):
    n_bins = int(lbp.max() + 1)
    return ax.hist(
        lbp.ravel(), density=True, bins=n_bins, range=(0, n_bins), facecolor='0.5'
    )

# plot histograms of LBP of textures
fig, (ax_img, ax_hist) = plt.subplots(nrows=2, ncols=3, figsize=(9, 6))
plt.gray()

titles = ('edge', 'flat', 'corner')
w = width = radius - 1
edge_labels = range(n_points // 2 - w, n_points // 2 + w + 1)
flat_labels = list(range(0, w + 1)) + list(range(n_points - w, n_points + 2))
i_14 = n_points // 4  # 1/4th of the histogram
i_34 = 3 * (n_points // 4)  # 3/4th of the histogram
corner_labels = list(range(i_14 - w, i_14 + w + 1)) + list(
    range(i_34 - w, i_34 + w + 1)
)

label_sets = (edge_labels, flat_labels, corner_labels)

for ax, labels in zip(ax_img, label_sets):
    ax.imshow(overlay_labels(image, lbp, labels))

for ax, labels, name in zip(ax_hist, label_sets, titles):
    counts, _, bars = hist(ax, lbp)
    highlight_bars(bars, labels)
    ax.set_ylim(top=np.max(counts[:-1]))
    ax.set_xlim(right=n_points + 2)
    ax.set_title(name)

ax_hist[0].set_ylabel('Percentage')
for ax in ax_img:
    ax.axis('off')
```

```python
METHOD = 'uniform'

def pick_bins(P, delta=None):
    δ = max(1, P // 12) if delta is None else delta  # smalere vindu = mer presis
    # FLAT: enten 0 eller P (og gjerne ignorere P+1 som er “ikke-uniform”)
    flat_bins = list(range(0, 0+δ+1)) + list(range(P-δ, P+1))
    # EDGE: rundt P/2
    mid = round(P/2)
    edge_bins = list(range(mid-δ, mid+δ+1))
    # CORNER: rundt P/4 og 3P/4
    q1, q3 = round(P/4), round(3*P/4)
    corner_bins = list(range(q1-δ, q1+δ+1)) + list(range(q3-δ, q3+δ+1))
    return flat_bins, edge_bins, corner_bins
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import local_binary_pattern
from skimage.filters import gaussian
from skimage.exposure import equalize_adapthist
from skimage.color import label2rgb
from PIL import Image

def imread_grayscale(path):
    return np.asarray(Image.open(path).convert("L"), dtype=np.float32) / 255.0

def lbp_hist(img, P, R, method='uniform'):
    lbp = local_binary_pattern(img, P, R, method)
    n_bins = int(lbp.max() + 1)
    hist, _ = np.histogram(lbp, bins=n_bins, range=(0, n_bins), density=True)
    return lbp, hist

def overlay_labels(image, lbp, labels):
    mask = np.isin(lbp, labels)
    return label2rgb(mask, image=image, bg_label=0, alpha=0.5)

def pick_bins(P, delta=None):
    δ = max(1, P // 12) if delta is None else delta
    flat_bins = list(range(0, 0+δ+1)) + list(range(P-δ, P+1))
    mid = round(P/2)
    edge_bins = list(range(mid-δ, mid+δ+1))
    q1, q3 = round(P/4), round(3*P/4)
    corner_bins = list(range(q1-δ, q1+δ+1)) + list(range(q3-δ, q3+δ+1))
    return flat_bins, edge_bins, corner_bins

# --- load & preprocess (Mona Lisa) ---
image = imread_grayscale("img/lbp/img1.png")
image_smooth = gaussian(image, sigma=1.0, preserve_range=True)
image_eq = equalize_adapthist(image_smooth, clip_limit=0.01)

# --- multiscale LBP ---
scales = [(8,1), (16,2), (24,3), (40,5)]  # (P, R)
lbp_maps, hists = [], []
for P, R in scales:
    lbp, h = lbp_hist(image_eq, P, R, method='uniform')
    lbp_maps.append((P,R,lbp))
    hists.append(h)

# (valgfritt) kombiner histogrammer for en global vurdering
H = np.concatenate(hists, axis=0)

# --- vis en valgt skala tydelig (f.eks. P=24, R=3) ---
P, R = 24, 3
lbp = [m for (PP,RR,m) in lbp_maps if PP==P and RR==R][0]
flat_bins, edge_bins, corner_bins = pick_bins(P, delta=2)

fig, axes = plt.subplots(2, 3, figsize=(10,7))
axes = axes.ravel()
axes[0].imshow(overlay_labels(image, lbp, edge_bins));   axes[0].set_title('Edges')
axes[1].imshow(overlay_labels(image, lbp, flat_bins));   axes[1].set_title('Flat')
axes[2].imshow(overlay_labels(image, lbp, corner_bins)); axes[2].set_title('Corners')
for i in range(3): axes[i].axis('off')

# histogram
n_bins = int(lbp.max()+1)
counts, bins, bars = axes[3].hist(lbp.ravel(), bins=n_bins, range=(0,n_bins), density=True)
axes[3].set_title('LBP histogram (P=24,R=3)')
axes[4].axis('off'); axes[5].axis('off')
plt.tight_layout()
```
