```python
outlier_cols = ['n_tokens_title', 'n_unique_tokens', 'n_non_stop_unique_tokens', 'num_hrefs', 'num_self_hrefs', 'num_imgs', 'num_videos', 'average_token_length', 'num_keywords']

Q1 = df[outlier_cols].quantile(0.25)
Q3 = df[outlier_cols].quantile(0.75)
IQR = Q3 - Q1

df_filtered = df[~((df[outlier_cols] < (Q1 - 5 * IQR)) | (df[outlier_cols] > (Q3 + 5 * IQR))).any(axis=1)]
```

```python
from sklearn.decomposition import PCA

pca = PCA()

df_pca = pd.DataFrame(pca.fit_transform(df))

df_pca
```

```python
plt.figure(figsize=(6, 4))
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_.cumsum(), label='Total Explained Variance')
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_, label='Explained Variance')
plt.title('Explained Variance vs Principal Components')
plt.ylabel('Explained Variance')
plt.xlabel('Principal Components')
plt.grid(visible=True)
plt.legend()
plt.show()
```

```python
import numpy as np

threshold = 0.95
pcs = np.argmax(pca.explained_variance_ratio_.cumsum() >= threshold) + 1

plt.figure(figsize=(6, 4))
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_.cumsum(), label='Total Explained Variance')
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_, label='Explained Variance')
plt.axhline(threshold, c='black', ls='--', label=f'{round(threshold*100)}% Threshold')
plt.axvline(pcs, c='black', ls='-', label=f'{pcs} Principal Components')
plt.title('Explained Variance vs Principal Components')
plt.ylabel('Explained Variance')
plt.xlabel('Principal Components')
plt.grid(visible=True)
plt.legend()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(df_pca, vars=df_pca.columns[:3], plot_kws={ 'alpha': 0.1, 's': 3 })
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df_pca[0], df_pca[1], df_pca[2], alpha=0.1, s=3)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

clusters = 4
kmeans = KMeans(n_clusters=clusters, n_init=20, random_state=1)
labels = kmeans.fit_predict(df_pca.iloc[:, :22])

sil_score = silhouette_score(df_pca.iloc[:, :3], labels)
db_score = davies_bouldin_score(df_pca.iloc[:, :3], labels)
```

```python
from matplotlib.lines import Line2D

cmap = plt.get_cmap('tab10', clusters)
colors = [cmap(i) for i in range(clusters)]

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for i in range(clusters):
    mask = labels == i 
    ax.scatter(df_pca[0][mask], df_pca[1][mask], df_pca[2][mask],
               color=colors[i], s=3, alpha=0.1)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

legend = [
    Line2D([0], [0], marker='o', color='w', label=f'Cluster {i+1}',
           markerfacecolor=colors[i], markersize=10)
    for i in range(clusters)
]
ax.legend(handles=legend, loc='upper right')

ax.text2D(0.05, 0.95, f"Silhouette Score: {sil_score:.3f}\nDavies-Bouldin Index: {db_score:.3f}", 
          transform=ax.transAxes, fontsize=10, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.6))

plt.show()
```

```python
loadings = pca.components_[:22].T
cluster_means = kmeans.cluster_centers_
feature_importance = np.abs(loadings @ cluster_means.T).sum(axis=1)

feature_importance_df = pd.DataFrame({
    'feature': df.columns,
    'importance': feature_importance
}).sort_values('importance', ascending=False)

top_features = feature_importance_df.head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='importance', y='feature', data=top_features, palette='viridis')
plt.title('Top 10 Features Contributing to K-means Clusters')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

cluster_range = range(3, 7)
sil_scores = []
db_scores = []

X = df_pca.iloc[:, :22].values

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, n_init=20, random_state=1)
    labels = kmeans.fit_predict(X)
    
    sil_scores.append(silhouette_score(df_pca.iloc[:, :3], labels))
    db_scores.append(davies_bouldin_score(df_pca.iloc[:, :3], labels))

plt.figure(figsize=(8,5))
plt.plot(cluster_range, sil_scores, label='Silhouette Score')
plt.plot(cluster_range, db_scores, label='Davies-Bouldin Index')
plt.xlabel('Clusters')
plt.ylabel('Score')
plt.xticks([3, 4, 5, 6])
plt.title('K-Means Clustering Scores')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from fcmeans import FCM
import numpy as np

clusters = 4
fcm = FCM(n_clusters=clusters, random_state=1)
fcm.fit(df_pca.iloc[:, :3].values)

labels = np.argmax(fcm.u, axis=1)

sil_score = silhouette_score(df_pca.iloc[:, :3], labels)
db_score = davies_bouldin_score(df_pca.iloc[:, :3], labels)
```

```python
alpha_vals = np.max(fcm.u, axis=1)

cmap = plt.get_cmap('tab10', clusters)
colors = [cmap(i) for i in range(clusters)]
cluster_colors = {i: colors[i] for i in range(clusters)}

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

for i in range(clusters):
    mask = labels == i
    ax.scatter(df_pca.iloc[mask, 0], df_pca.iloc[mask, 1], df_pca.iloc[mask, 2],
               color=cluster_colors[i], s=3, alpha=0.01 + 0.09*alpha_vals[mask])

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

legend_elements = [
    Line2D([0], [0], marker='o', color='w', label=f'Cluster {i+1}',
           markerfacecolor=cluster_colors[i], markersize=10)
    for i in range(clusters)
]
ax.legend(handles=legend_elements, loc='upper right')

ax.text2D(0.05, 0.95, f"Silhouette Score: {sil_score:.3f}\nDavies-Bouldin Index: {db_score:.3f}", 
          transform=ax.transAxes, fontsize=10, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.6))

plt.show()
```

```python
loadings = pca.components_[:3].T
cluster_means = fcm.centers
feature_importance = np.abs(loadings @ cluster_means.T).sum(axis=1)

feature_importance_df = pd.DataFrame({
    'feature': df.columns,
    'importance': feature_importance
}).sort_values('importance', ascending=False)

top_features = feature_importance_df.head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='importance', y='feature', data=top_features, palette='viridis')
plt.title('Top 10 Features Contributing to Fuzzy C-means Clusters')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```
