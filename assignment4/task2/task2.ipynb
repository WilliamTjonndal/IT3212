```python
import pandas as pd

df = pd.read_csv('social-media-dataset/OnlineNewsPopularity.csv')

df.head()
```

```python
df.drop(columns=['url', 'weekday_is_monday', 'weekday_is_tuesday', 'weekday_is_wednesday', 'weekday_is_thursday', 'weekday_is_friday', 'weekday_is_saturday', 'weekday_is_sunday', 'is_weekend', 'timedelta', 'shares'], inplace=True)

df.head()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

df_scaled
```

```python
df = df_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA()

df_pca = pd.DataFrame(pca.fit_transform(df))

df_pca
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_.cumsum(), label='Total Explained Variance')
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_, label='Explained Variance')
plt.title('Explained Variance vs Principal Components')
plt.ylabel('Explained Variance')
plt.xlabel('Principal Components')
plt.grid(visible=True)
plt.legend()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(df_pca, vars=df_pca.columns[:5])
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df_pca[0], df_pca[1], df_pca[2], alpha=0.1)

ax.set_xlabel('PC1', rotation=-18)
ax.set_ylabel('PC2', rotation=54)
ax.set_zlabel('PC3', rotation=90)

plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import numpy as np

clusters = 4
kmeans = KMeans(n_clusters=clusters, random_state=1)
labels = kmeans.fit_predict(df_pca)

cmap = cm.get_cmap('coolwarm', clusters) 

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')

for cluster_id in range(clusters):
    mask = labels == cluster_id
    ax.scatter(df_pca[0][mask], df_pca[1][mask], df_pca[2][mask],
               c=np.array([cmap(cluster_id)]), label=f'Cluster {cluster_id}', s=3, alpha=0.1)
    
for i, center in enumerate(kmeans.cluster_centers_):
    ax.scatter(center[0], center[1], center[2], c='black', s=150, marker='X', edgecolor='white', linewidth=1.5)
    ax.text(center[0], center[1], center[2], f'Cluster {i}', color='black', fontsize=10)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

plt.legend()
plt.show()
```
