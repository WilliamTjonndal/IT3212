```python
import pandas as pd

df = pd.read_csv('social-media-dataset/OnlineNewsPopularity.csv')

df.head()
```

```python
df.drop(columns=['url', 'weekday_is_monday', 'weekday_is_tuesday', 'weekday_is_wednesday', 'weekday_is_thursday', 'weekday_is_friday', 'weekday_is_saturday', 'weekday_is_sunday', 'is_weekend', 'timedelta'], inplace=True)

df.head()
```

```python
df.info()
```

```python
pd.set_option('display.expand_frame_repr', True)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

df.describe(include='all')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 4))
df.boxplot(rot=90, grid=False)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

df_scaled
```

```python
df = df_scaled
```

```python
df
```

```python
plt.figure(figsize=(16, 4))
df.boxplot(rot=90, grid=False)
plt.show()
```

```python
cols_to_check = ['n_tokens_title', 'n_tokens_content', 'n_unique_tokens', 'n_non_stop_unique_tokens', 'num_hrefs', 'average_token_length', 'num_keywords']

Q1 = df[cols_to_check].quantile(0.25)
Q3 = df[cols_to_check].quantile(0.75)
IQR = Q3 - Q1

df_filtered = df[~((df[cols_to_check] < (Q1 - 3 * IQR)) | (df[cols_to_check] > (Q3 + 3 * IQR))).any(axis=1)]
```

```python
df = df_filtered
```

```python
plt.figure(figsize=(16, 4))
df.boxplot(rot=90, grid=False)
plt.show()
```

```python
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

df_scaled
```

```python
df = df_scaled
```

```python
plt.figure(figsize=(16, 4))
df.boxplot(rot=90, grid=False)
plt.show()
```

```python
df
```

```python
from sklearn.decomposition import PCA

pca = PCA()

df_pca = pd.DataFrame(pca.fit_transform(df))

df_pca
```

```python
plt.figure(figsize=(6, 4))
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_.cumsum(), label='Total Explained Variance')
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_, label='Explained Variance')
plt.title('Explained Variance vs Principal Components')
plt.ylabel('Explained Variance')
plt.xlabel('Principal Components')
plt.grid(visible=True)
plt.legend()
plt.show()
```

```python
import numpy as np

threshold = 0.95
pcs = np.argmax(pca.explained_variance_ratio_.cumsum() >= threshold) + 1

plt.figure(figsize=(6, 4))
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_.cumsum(), label='Total Explained Variance')
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_, label='Explained Variance')
plt.axhline(threshold, c='black', ls='--', label=f'{round(threshold*100)}% Threshold')
plt.axvline(pcs, c='black', ls='-', label=f'{pcs} Principal Components')
plt.title('Explained Variance vs Principal Components')
plt.ylabel('Explained Variance')
plt.xlabel('Principal Components')
plt.grid(visible=True)
plt.legend()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(df_pca, vars=df_pca.columns[:3], plot_kws={ 'alpha': 0.1, 's': 3 })
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df_pca[0], df_pca[1], df_pca[2], alpha=0.1, s=3)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

plt.show()
```

```python
from sklearn.cluster import KMeans

clusters = 6
kmeans = KMeans(n_clusters=clusters, n_init=20, random_state=1)
labels = kmeans.fit_predict(df_pca.iloc[:, :22])
```

```python
from matplotlib.lines import Line2D

cmap = plt.get_cmap('tab10', clusters)
colors = [cmap(i) for i in range(clusters)]

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for i in range(clusters):
    mask = labels == i 
    ax.scatter(df_pca[0][mask], df_pca[1][mask], df_pca[2][mask],
               color=colors[i], s=3, alpha=0.1)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

legend = [
    Line2D([0], [0], marker='o', color='w', label=f'Cluster {i+1}',
           markerfacecolor=colors[i], markersize=10)
    for i in range(clusters)
]
ax.legend(handles=legend, loc='upper right')

plt.show()
```

```python
loadings = pca.components_[:22].T
cluster_means = kmeans.cluster_centers_
feature_importance = np.abs(loadings @ cluster_means.T).sum(axis=1)

feature_importance_df = pd.DataFrame({
    'feature': df.columns,
    'importance': feature_importance
}).sort_values('importance', ascending=False)

top_features = feature_importance_df.head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='importance', y='feature', data=top_features, palette='viridis')
plt.title('Top 10 Features Contributing to K-means Clusters')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
from fcmeans import FCM
import numpy as np

clusters = 4
fcm = FCM(n_clusters=clusters, random_state=1)
fcm.fit(df_pca.iloc[:, :3].values)

labels = np.argmax(fcm.u, axis=1)
```

```python
alpha_vals = np.max(fcm.u, axis=1)

cmap = plt.get_cmap('tab10', clusters)
colors = [cmap(i) for i in range(clusters)]
cluster_colors = {i: colors[i] for i in range(clusters)}

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

for i in range(clusters):
    mask = labels == i
    ax.scatter(df_pca.iloc[mask, 0], df_pca.iloc[mask, 1], df_pca.iloc[mask, 2],
               color=cluster_colors[i], s=3, alpha=0.01 + 0.09*alpha_vals[mask])

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

legend_elements = [
    Line2D([0], [0], marker='o', color='w', label=f'Cluster {i+1}',
           markerfacecolor=cluster_colors[i], markersize=10)
    for i in range(clusters)
]

ax.legend(handles=legend_elements, loc='upper right')
plt.show()
```

```python
loadings = pca.components_[:3].T
cluster_means = fcm.centers
feature_importance = np.abs(loadings @ cluster_means.T).sum(axis=1)

feature_importance_df = pd.DataFrame({
    'feature': df.columns,
    'importance': feature_importance
}).sort_values('importance', ascending=False)

top_features = feature_importance_df.head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='importance', y='feature', data=top_features, palette='viridis')
plt.title('Top 10 Features Contributing to Fuzzy C-means Clusters')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
from sklearn.mixture import GaussianMixture

clusters = 4
gmm = GaussianMixture(n_components=clusters, n_init=20, random_state=1)
labels = gmm.fit_predict(df_pca.iloc[:, :22])
```

```python
from matplotlib.lines import Line2D

cmap = plt.get_cmap('tab10', clusters)
colors = [cmap(i) for i in range(clusters)]

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for i in range(clusters):
    mask = labels == i 
    ax.scatter(df_pca[0][mask], df_pca[1][mask], df_pca[2][mask],
               color=colors[i], s=3, alpha=0.1)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

legend = [
    Line2D([0], [0], marker='o', color='w', label=f'Cluster {i+1}',
           markerfacecolor=colors[i], markersize=10)
    for i in range(clusters)
]
ax.legend(handles=legend, loc='upper right')

plt.show()
```

```python
loadings = pca.components_[:22].T

cluster_means = gmm.means_
feature_importance = np.abs(loadings @ cluster_means.T).sum(axis=1)

feature_importance_df = pd.DataFrame({
    'feature': df.columns,
    'importance': feature_importance
}).sort_values('importance', ascending=False)

top_features = feature_importance_df.head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='importance', y='feature', data=top_features, palette='viridis')
plt.title('Top 10 Features Contributing to GMM Clusters')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```
