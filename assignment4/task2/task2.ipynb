```python
import pandas as pd

df = pd.read_csv('social-media-dataset/OnlineNewsPopularity.csv')

df.head()
```

```python
df.drop(columns=['url', 'weekday_is_monday', 'weekday_is_tuesday', 'weekday_is_wednesday', 'weekday_is_thursday', 'weekday_is_friday', 'weekday_is_saturday', 'weekday_is_sunday', 'is_weekend', 'timedelta', 'shares'], inplace=True)

df.head()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

df_scaled
```

```python
df = df_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA()

df_pca = pd.DataFrame(pca.fit_transform(df))

df_pca
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_.cumsum(), label='Total Explained Variance')
plt.plot(range(1, len(pca.components_) + 1), pca.explained_variance_ratio_, label='Explained Variance')
plt.title('Explained Variance vs Principal Components')
plt.ylabel('Explained Variance')
plt.xlabel('Principal Components')
plt.grid(visible=True)
plt.legend()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(df_pca, vars=df_pca.columns[:5], plot_kws={ 'alpha': 0.1, 's': 3 })
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df_pca[0], df_pca[1], df_pca[2], alpha=0.1, s=3)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

plt.show()
```

```python
from sklearn.cluster import KMeans

clusters = 3
kmeans = KMeans(n_clusters=clusters, n_init=20, random_state=1)
labels = kmeans.fit_predict(df_pca)
```

```python
from matplotlib.lines import Line2D

cmap = plt.get_cmap('tab10', clusters)
colors = [cmap(i) for i in range(clusters)]

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for i in range(clusters):
    mask = labels == i 
    ax.scatter(df_pca[0][mask], df_pca[1][mask], df_pca[2][mask],
               color=colors[i], s=3, alpha=0.1)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

legend = [
    Line2D([0], [0], marker='o', color='w', label=f'Cluster {i+1}',
           markerfacecolor=colors[i], markersize=10)
    for i in range(clusters)
]
ax.legend(handles=legend, loc='upper right')

plt.show()
```

```python
from hdbscan import HDBSCAN

dbscan = HDBSCAN(min_cluster_size=1000)
labels = dbscan.fit_predict(df_pca)
```

```python
from matplotlib.lines import Line2D
import matplotlib.pyplot as plt

# --- Color assignment ---
unique_labels = sorted(set(labels))
clusters = len(unique_labels) - (1 if -1 in labels else 0)

cmap = plt.get_cmap('tab10', clusters)
colors = [cmap(i) for i in range(clusters)]

cluster_colors = {}
color_idx = 0
for lab in unique_labels:
    if lab == -1:
        cluster_colors[lab] = (0.5, 0.5, 0.5, 0.3)  # gray for noise
    else:
        cluster_colors[lab] = colors[color_idx]
        color_idx += 1

# --- 3D Scatter plot ---
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

for lab in unique_labels:
    mask = labels == lab
    alpha_val = 0.1 if lab != -1 else 0.05
    s_val = 5 if lab != -1 else 3
    ax.scatter(df_pca.iloc[mask, 0], df_pca.iloc[mask, 1], df_pca.iloc[mask, 2],
               color=cluster_colors[lab], s=s_val, alpha=alpha_val)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

# --- Legend ---
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label=f'Cluster {lab+1}' if lab != -1 else 'Noise',
           markerfacecolor=cluster_colors[lab], markersize=10)
    for lab in unique_labels
]

ax.legend(handles=legend_elements, loc='upper right')
plt.show()
```

```python
from fcmeans import FCM
import numpy as np

clusters = 3
fcm = FCM(n_clusters=clusters, random_state=1)
fcm.fit(df_pca.iloc[:, :3].values)

labels = np.argmax(fcm.u, axis=1)
```

```python
alpha_vals = np.max(fcm.u, axis=1)

cmap = plt.get_cmap('tab10', clusters)
colors = [cmap(i) for i in range(clusters)]
cluster_colors = {i: colors[i] for i in range(clusters)}

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

for i in range(clusters):
    mask = labels == i
    ax.scatter(df_pca.iloc[mask, 0], df_pca.iloc[mask, 1], df_pca.iloc[mask, 2],
               color=cluster_colors[i], s=3, alpha=0.01 + 0.09*alpha_vals[mask])

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

legend_elements = [
    Line2D([0], [0], marker='o', color='w', label=f'Cluster {i+1}',
           markerfacecolor=cluster_colors[i], markersize=10)
    for i in range(clusters)
]

ax.legend(handles=legend_elements, loc='upper right')
plt.show()
```
